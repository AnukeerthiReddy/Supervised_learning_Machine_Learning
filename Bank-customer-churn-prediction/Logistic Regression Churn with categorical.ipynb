```python
data = pd.read_csv("Bank Customer Churn Prediction.csv")
```

```python
import scipy.stats as stats
import statistics

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
import xgboost as xgb
from sklearn.metrics import accuracy_score, precision_score, recall_score,classification_report,confusion_matrix
```

```python
#dropping the catrgorical columns as well for model deployment without transformation and scaling
data.drop(columns=['customer_id'], inplace=True)
data.head()
```

```python
data.head()
```

```python
categorical_columns = ['country','gender']
# encoder = pd.get_dummies()
# encoder.fit(X_train_categorical)
# Perform one-hot encoding
one_hot_encoded = pd.get_dummies(data[categorical_columns])
# Concatenate the one-hot encoded columns to the original dataframe
data = pd.concat([data, one_hot_encoded], axis=1)
#Drop original categorical columns
data.drop(categorical_columns,axis=1,inplace=True)
```

```python
data.head()
```

```python
#Train Test Split
X = data.iloc[:, data.columns != data.columns[8]]
y = data.iloc[:,8]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=17,stratify=y)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
classifier = LogisticRegression()
classifier.fit(X,y)
```

```python
y_pred = classifier.predict(X)
```

```python
y_pred_df= pd.DataFrame({'actual': y,
                         'predicted_prob': classifier.predict(X)})
```

```python
y_pred_df
```

```python
# Confusion Matrix for the model accuracy
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y,y_pred)
print (confusion_matrix)
```

```python
#Classification report
from sklearn.metrics import classification_report
print(classification_report(y,y_pred))
```
